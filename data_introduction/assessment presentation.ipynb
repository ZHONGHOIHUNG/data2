```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from datetime import datetime
%matplotlib inline
filepath = "1.csv" 


penalty_data = pd.read_csv(filepath)
print("done")
penalty_data.head(100)
```

```python
def make_date(year):
    try:
        return datetime(year=int(year), month=6, day=1)
    except:
        print("bad year data:", year, type(year))
        return None
penalty_data["year_dt"] = penalty_data.Year.apply(make_date)
fig, ax = plt.subplots()
for country, df in penalty_data.set_index("year_dt").groupby("Country or Area"):
    df.Value.plot()
    try:
        ax.annotate(
            country,
            xy=(df.index.max(), df.Value[pd.Series(df.index).idxmax()]),
            xytext=(1, 0), textcoords='offset points',
        )
    except:
        pass
```
